```python
# Importing chatterbot
from chatterbot import ChatBot
```

```python
# Create object of ChatBot class
bot = ChatBot('ServiceAgent')
```

```python
# Create object of ChatBot class with Storage Adapter
bot = ChatBot(
    'ServiceAgent',
    storage_adapter='chatterbot.storage.SQLStorageAdapter',
    database_uri='sqlite:///database.sqlite3'
)
```

```python
# Create object of ChatBot class with Logic Adapter
bot = ChatBot(
    'ServiceAgent',  
    logic_adapters=[
        'chatterbot.logic.BestMatch',
        'chatterbot.logic.TimeLogicAdapter']
)
# The TimeLogicAdapter auto identifies numbers as time, as there isn't any other code that handles data types outside of strings
```

```python
# https://chatterbot.readthedocs.io/en/latest/training.html#training-with-corpus-data

# Inport ListTrainer
from chatterbot.trainers import ListTrainer

trainer = ListTrainer(bot)

trainer.train([
'Hi',
'Hello',
'I need your assistance regarding my order',
'Please, Provide me with your order id, beginning with "my order id is"',
'my order id is ',
'Thanks for that information. How can I help you?',
'I have a complaint.',
'Please elaborate, your concern',
'How long it will take to receive an order ?',
'An order takes 3-5 Business days to get delivered.',
'Okay Thanks',
'No Problem! Have a Good Day!'
])

#additional training options below:
# from chatterbot.trainers import ChatterBotCorpusTrainer

# trainer = ChatterBotCorpusTrainer(bot)

# trainer.train(
#    "chatterbot.corpus.english"

```

```python
# Get a response to the input text 'I would like to book a flight.'
response = bot.get_response('I have a complaint.')

print("Bot Response:", response)
```
